#Task 1

```python

food = 500
transport = 150
stationery = 300

total_expense = food + transport + stationery

budget = 1500

print("Total daily expense:", total_expense, "PKR")

if total_expense > budget:
    print("Budget exceeded!")
else:
    print("You are within the budget.")
```

#Task 3

```python

temperatures = [18, 25, 32, 28, 35, 14, 30, 40]

normal_count = 0
high_count = 0

for temp in temperatures:
    if 15 <= temp <= 30:
        normal_count += 1
    elif temp > 30:
        high_count += 1

print("Normal temperature readings:", normal_count)
print("High temperature readings:", high_count)
```

#Task 4

```python

incomes = [75000, 50000, 62000, 80000, 45000]
employment_status = ["Employed", "Employed", "Unemployed", "Employed", "Employed"]

approved_count = 0

for i in range(len(incomes)):
    if incomes[i] >= 60000 and employment_status[i] == "Employed":
        approved_count += 1

print("Number of applicants approved for loan:", approved_count)
```

#Task 5

```python

speeds = [60, 85, 90, 75, 100, 80, 95]
violation_count = 0
total_vehicles = len(speeds)
for speed in speeds:
    if speed > 80:
        violation_count += 1


percentage = (violation_count / total_vehicles) * 100
print("Total vehicles:", total_vehicles)
print("Overspeeding vehicles:", violation_count)
print("Violation percentage:", percentage, "%")
```
