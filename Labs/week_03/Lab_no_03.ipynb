TASK #01

```python
def remove_invalid_readings(data):
    valid_data = []
    for value in data:
        if 0 <= value <= 100:
            valid_data.append(value)
    return valid_data

def calculate_average(data):
    if len(data) == 0:
        return 0
    return sum(data) / len(data)

sensor_readings = [45, 78, -12, 90, 105, 66, 88]

cleaned_data = remove_invalid_readings(sensor_readings)
average = calculate_average(cleaned_data)

print("Original Data:", sensor_readings)
print("Cleaned Data:", cleaned_data)
print("Average of Cleaned Data:", average)
```

TASK# 02

```python
def calculate_average(marks):
    if len(marks) == 0:
        return 0
    else:
        return sum(marks) / len(marks)

def check_pass_status(average):
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

def display_summary(students):
    for name, marks in students.items():
        avg = calculate_average(marks)
        status = check_pass_status(avg)

        print("Student:", name)
        print("Marks:", marks)
        print("Average:", round(avg, 2))
        print("Status:", status)
        print("-" * 30)

students = {
    "Ali": [65, 70, 80],
    "Sara": [90, 85, 88],
    "Ahmed": [40, 55, 60]
}

display_summary(students)
```

TASK #03

```python
class NumericDataset:
    def __init__(self, values):
        self.values = values

    def count_data_points(self):
        if len(self.values) == 0:
            return 0
        else:
            return len(self.values)

    def calculate_average(self):
        if len(self.values) == 0:
            return 0
        else:
            return sum(self.values) / len(self.values)


data_values = [12, 18, 25, 30, 22, 27]

dataset = NumericDataset(data_values)

print("Data Values:", dataset.values)
print("Number of Data Points:", dataset.count_data_points())
print("Average:", dataset.calculate_average())
```

TASK #04

```python
class SimpleClassifier:
    def __init__(self, threshold):
        self.threshold = threshold

    def classify_value(self, value):
        if value >= self.threshold:
            return True
        else:
            return False

    def classify_list(self, values):
        classifications = []
        for value in values:
            classification = self.classify_value(value)
            classifications.append(classification)
        return classifications


threshold = 60
values = [45, 72, 88, 30, 65]

classifier = SimpleClassifier(threshold)

print("Threshold:", classifier.threshold)
print("Values:", values)
print("List Classification:", classifier.classify_list(values))
```
